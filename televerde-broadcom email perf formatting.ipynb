```python
import pandas as pd
import os

# Read the CSV file as a DataFrame
file_path = 'C:/Users/LuqmanJunaida/Desktop/broadcom_rally_22-06-2023.csv'
file_name = os.path.splitext(os.path.basename(file_path))[0]  # Extract the file name without extension
df = pd.read_csv(file_path, header=None, dtype=str)

# Extract the header row
header_row = df.iloc[1]

# Set the header row as the column names
df.columns = header_row

# Remove the first two rows (header row and the original first row)
df = df.iloc[2:]

# Save 'Contact Info' data
contact_info_df = df.iloc[:, :14].copy()  # Make a copy to avoid the SettingWithCopyWarning

# Initialize an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Process 'Email 1' to 'Email 6' data
for i in range(1, 7):
    email_start_column = 14 + (i - 1) * 21  # Calculate the starting column index for the current email
    email_end_column = email_start_column + 21  # Calculate the ending column index for the current email
    email_df = df.iloc[:, email_start_column:email_end_column].copy()  # Extract the columns for the current email and make a copy
    email_df.columns = email_df.columns.astype(str)  # Convert column names to string type
    
    # Append 'Bounced' column to contact_info_df
    contact_info_df['Bounced'] = email_df['Bounced']
    
    variations = ['v1', 'v2', 'v3', 'v4']
    combined_email_df = pd.DataFrame()

    for variation in variations:
        variation_columns = [col for col in email_df.columns if col.endswith(variation)]
        variation_data = email_df[variation_columns]
        variation_data.columns = [col.split()[0] for col in variation_columns]
        variation_data['Variation'] = 'V' + variation[1]
        combined_variation_df = pd.concat([contact_info_df, variation_data], axis=1)
        combined_email_df = pd.concat([combined_email_df, combined_variation_df], axis=0)
    
    # Insert 'Bounced' column back into combined DataFrame
    combined_email_df['Bounced'] = contact_info_df['Bounced']
    
    email_df['Email'] = 'Email ' + str(i)
    combined_email_df['Email'] = 'Email ' + str(i)
    combined_email_df = pd.concat([combined_email_df, email_df], axis=0)
    combined_df = pd.concat([combined_df, combined_email_df], axis=0)

# Remove rows where 'Variation' column is blank
combined_df = combined_df[combined_df['Variation'].notnull()]

# Determine the Email Campaign Name based on the file name
if 'rally' in file_name:
    email_campaign_name = 'Broadcom Rally Starter Edition'
elif 'lead pool' in file_name:
    email_campaign_name = 'Broadcom Lead Pool Nurture'
else:
    email_campaign_name = ''

# Add 'Email Campaign Name' column and assign the determined value
combined_df['Email Campaign Name'] = email_campaign_name

# Add 'Email Name & Number' column by concatenating 'Email Campaign Name' and 'Email' column
combined_df['Email Name & Number'] = combined_df['Email Campaign Name'] + ' ' + combined_df['Email']

# Select the desired columns
selected_columns = ['Job Title', 'Company Name', 'Country', 'Industry', 'Num Employees Range', 'Revenue Range',
                    'Person Score', 'Ranking Type', 'Original Responder Type', 'Original Responder Detail',
                   'Most Recent Responder Detail', 'Email Address', 'Solution Area', 'NLWC', 'Sent', 'Opened',
                    'Clicked', 'Unsubscribed', 'Delivered', 'Bounced', 'Email', 'Variation', 'Email Campaign Name',
                    'Email Name & Number']
                
#selected_columns = ['Job Title', 'Company Name', 'Country', 'Industry', 'Num Employees Range', 'Revenue Range',
#                    'Person Score', 'Ranking Type', 'Email Address', 'Solution Area', 'Sent', 'Opened',
#                    'Clicked', 'Unsubscribed', 'Delivered', 'Bounced', 'Email', 'Variation', 'Email Campaign Name',
#                    'Email Name & Number']                

# Filter the combined DataFrame with the selected columns
combined_df = combined_df[selected_columns]

# Save the combined DataFrame to a CSV file with the modified file name
output_file_path = f'C:/Users/LuqmanJunaida/Downloads/{file_name}_output.csv'
combined_df.to_csv(output_file_path, index=False)

print("CSV file saved successfully.")
```
